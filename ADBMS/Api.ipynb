```python
import requests
import pandas as pd
```

```python
url=  "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
data = response.json()
print("Data Extracted Sucessfully")
```

```python
df=pd.DataFrame(data)
df.head()
```
