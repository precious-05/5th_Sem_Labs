```python
import requests
import pandas as pd

api_url = "665f1f975fe5874e657815c39ad0d901"
url2= "https://api.openweathermap.org/data/2.5/weather?lat=30.9644&lon=70.9360&appid=665f1f975fe5874e657815c39ad0d901"
```

```python
response = requests.get(url2)
print(response)
```

```python
data = response.json()
```

```python
print(data)
```

```python
#----------------------------DataTransformation-----------------------------
transformed_data = {
    "City": data["name"],
    "Temperature_C": data["main"]["temp"],
    "Feels_Like_C": data["main"]["feels_like"],
    "Humidity": data["main"]["humidity"],
    "Pressure": data["main"]["pressure"],
    "Weather_Main": data["weather"][0]["main"],
    "Weather_Description": data["weather"][0]["description"],
    "Wind_Speed": data["wind"]["speed"]
}

```

```python
#---------------------Converting dictionary to DataFrame--------------------------------
df = pd.DataFrame([transformed_data])

# ---------- LOADING DATA (SAVING TO CSV FILE) ----------
df.to_csv("weather_data.csv", index=False)

print("ETL completed successfully. The CSV file saved as weather_data.csv")
print(df)
```
