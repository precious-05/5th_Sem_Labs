```python
import requests
import pandas as pd

# --- Extract ---
GITHUB_TOKEN = "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"
HEADERS = {"Authorization": f"token {GITHUB_TOKEN}"}

def extract_github_data(api_url):
    data = []
    page = 1
    while True:
        params = {"page": page, "per_page": 100}
        response = requests.get(api_url, headers=HEADERS, params=params)
        response.raise_for_status()
        page_data = response.json()
        if not page_data:
            break
        data.extend(page_data)
        page += 1
    return data

user_repos_url = "https://api.github.com/users/precious-05/repos"
repositories = extract_github_data(user_repos_url)

print(f"Total repositories fetched: {len(repositories)}")

# --- Transform ---
# Ab hum sirf kuch relevant columns nikalte hain jo analysis me useful hon:
filtered_data = []
for repo in repositories:
    filtered_repo = {
        "name": repo.get("name"),
        "full_name": repo.get("full_name"),
        "private": repo.get("private"),
        "html_url": repo.get("html_url"),
        "description": repo.get("description"),
        "fork": repo.get("fork"),
        "created_at": repo.get("created_at"),
        "updated_at": repo.get("updated_at"),
        "pushed_at": repo.get("pushed_at"),
        "stargazers_count": repo.get("stargazers_count"),
        "watchers_count": repo.get("watchers_count"),
        "language": repo.get("language"),
        "forks_count": repo.get("forks_count"),
        "open_issues_count": repo.get("open_issues_count"),
        "default_branch": repo.get("default_branch")
    }
    filtered_data.append(filtered_repo)

# Convert filtered data to DataFrame
df = pd.DataFrame(filtered_data)

# --- Load ---
# CSV file me save kar dete hain
csv_file_path = "github_repositories.csv"
df.to_csv(csv_file_path, index=False)

print(f"Data saved to {csv_file_path}")

# Agar chaho to CSV ka pehla 5 rows print karwa lo confirm karne ke liye
print(df.head())
```
